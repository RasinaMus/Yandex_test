```python
# проверим order_price на NaN
df.dropna(subset=['order_price'], how='all')
```

```python
# Размерности совпадают, значит пропусков в данных в сумме заказов нет
```

```python
# оставим те заказы, где сумма заказа не равна нулю
df_2 = df[df['order_price'] != 0]
df_2
```

```python
# добавим в таблицу столбец с процентным соотношением количества чаевых к сумме заказа
df_2['pct'] = (np.array(df_2['tips']) / np.array(df_2['order_price'])) * 100
df_2
```

## Простая проверка. Возьмем среднее значение чаевых в процентах тех, у кого было меньше трех комплектов приборов и больше двух и сравним

```python
# у кого меньше трех комплектов
cultery1 = np.array(df_2[df_2['cutlery'] < 3]['pct'])
cultery1.mean()
```

```python
# у кого больше двух комплектов
cultery2 = np.array(df_2[df_2['cutlery'] > 2]['pct'])
cultery2.mean()
```

#### Среднее значение чаевых у тех, кто брал больше двух комплектов: 12.338199448567519; у кого меньше трех: 8.833760753392083, значит гипотеза верна

## Еще проверка

### Посмотрим на 30 процентов самых больших чевых и среди них сравним количество человек с набором приборов меньше трех и больше двух

```python
quant = df_2['pct'].quantile(0.7)
```

```python
num1 = df_2[df_2['pct'] > quant]
num1[num1['cutlery'] < 3].shape[0]
```

```python
num1[num1['cutlery'] > 2].shape[0]
```

#### Количество человек с тремя наборами и более среди 30 процентов самых больших чаевых: 5308, а тех, у кого два и один набор: 1682. Тогда гипотеза опять подтверждается

```python
dates_to_del = ['01', '02', '03', '04', '05', '06', '07', '08', '09'] # дни января, которые не нужно учитывать
```

```python
df_new = df_2[df_2['cutlery'] > 2]
df_new = df_new[df_new['order_price'] > 800]
df_new
```

```python
df_new['date'].loc[0].split(' ')[0].split('-')[-1]
```

```python
l = []
for i in list(df_new.index):
    if df_new['date'].loc[i].split(' ')[0].split('-')[1] == '01' and df_new['date'].loc[i].split(' ')[0].split('-')[-1] in dates_to_del:
        l.append(i)
len(l)
```

```python
df_all = df_new.loc[set(df_new.index) - set(l)]
df_all
```
